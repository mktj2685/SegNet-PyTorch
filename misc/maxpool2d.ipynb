# Check `MaxPool()` & `UnMaxPool()`

```python
import seaborn as sns
import torch
import torch.nn as nn
torch.cuda.is_available()
```

```python
x = torch.randint(-10, 10, (4, 4)).float()    # dummy feature plane : (H, W)
sns.heatmap(x, annot=True)
```

```python
pool = nn.MaxPool2d(kernel_size=2, stride=1, return_indices=True)
x = x[None, None, :, :]     # dummy feature map : (B, C, H, W)
y, indices = pool(x)
sns.heatmap(y[0,0,:,:], annot=True)
```

```python
unpool = nn.MaxUnpool2d(kernel_size=2, stride=1)
z = unpool(y, indices)
sns.heatmap(z[0,0,:,:], annot=True)
```
